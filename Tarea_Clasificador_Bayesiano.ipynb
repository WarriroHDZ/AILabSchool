Prioriti:

```python
prioritiSetosa=probabilidadLargoSepaloSetosa*probabilidadAnchoSepaloSetosa*probabilidadLargoPetaloSetosa*probabilidadAnchoPetaloSetosa
prioritiVersicolor=probabilidadLargoSepaloVersicolor*probabilidadAnchoSepaloVersicolor*probabilidadLargoPetaloVersicolor*probabilidadAnchoPetaloVersicolor
prioritiVirginica=probabilidadLargoSepaloVirginica*probabilidadAnchoSepaloVirginica*probabilidadLargoPetaloVirginica*probabilidadAnchoPetaloVirginica
```

Evidencia:

```python
evidencia=prioritiSetosa+prioritiVersicolor+prioritiVirginica
```

Posteriori

```python
posterioriSetosa=prioritiSetosa/evidencia
posterioriVersicolor=prioritiVersicolor/evidencia
posterioriVirginica=prioritiVirginica/evidencia
```

Sentencia Final:

```python
if (posterioriSetosa > posterioriVersicolor and posterioriVirginica):
    print("Es una Setosa")
elif (posterioriVersicolor > posterioriVirginica and posterioriSetosa):
    print("Es una Versicolor")
elif (posterioriVirginica > posterioriSetosa and posterioriVirginica):
    print("Es una Virginica")
else:
    print("No se puede clasificar")
```
